Analysis of data taken on MD shift on 24/02/2026

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib ipympl
```

```python
df = pd.read_csv('2026-02-24__16-33-07.csv')
```

```python
x = range(df.shape[0])

fig, ax = plt.subplots(2, 1, figsize = (7, 10))
ax[0].plot(x, df['Add (ID: 38)'])
ax[0].axvline(14, ls = '--', alpha = .5)
ax[0].text(6, 3, 'Random Samples', rotation = 90, fontsize = 8)
ax[0].set_xlim(0, x[-1])
ax[0].set_xlabel('Evaluation')
ax[0].set_ylabel('Objective')
ax[1].plot(x, np.maximum.accumulate(df['Add (ID: 38)']))
ax[1].axvline(14, ls = '--', alpha = .5)
ax[1].text(6, 3, 'Random Samples', rotation = 90, fontsize = 8)
ax[1].set_xlim(0, x[-1])
ax[1].set_xlabel('Evaluation')
ax[1].set_ylabel('Objective')
fig.savefig('C:\\Users\\shaun\\fig.png', dpi = 300, bbox_inches = 'tight')
```
