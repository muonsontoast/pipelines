# Analysis of GP XOpt Output

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib ipympl
```

```python
df = pd.read_csv(f'{os.getcwd()}\\GP-run-1.txt', sep = ' ')

fig, ax = plt.subplots(figsize = (5, 5))
ax.scatter(df['34-0'], df['34-1'])
ax.grid(alpha = .35)
```

```python
df = pd.read_csv(f'{os.getcwd()}\\GP-run-2.txt', sep = ' ')

fig, ax = plt.subplots(figsize = (5, 5))
ax.scatter(df['34-0'], df['34-1'])
ax.grid(alpha = .35)
```

```python
df = pd.read_csv(f'{os.getcwd()}\\GP-run-3.txt', sep = ' ')

fig, ax = plt.subplots(figsize = (5.5, 5))
ax.scatter(df['34-0'], df['34-1'], s = 15, marker = '*', label = 'beam lost')
survivors = np.array([
    [df['34-0'][it], df['34-1'][it]] for it in range(len(df)) if df['f'][it] == 0
])
ax.scatter(survivors[:, 0], survivors[:, 1], s = 15, marker = '*', color = 'tab:red', label = 'beam survives')
ax.grid(alpha = .35)
ax.set_xlabel('Singular Value 1')
ax.set_ylabel('Singular Value 2')
ax.legend(loc = 'lower center')
ax.set_title('7 HSTR + 7 VSTR BTS Optimisation (Simulation)')
fig.savefig('GP-run-3.png', dpi = 300, bbox_inches = 'tight')
```
